```python
import numpy as np
import cv2
import os
import tensorflow as tf
from sklearn.utils import shuffle
from sklearn.feature_extraction import image
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
```

```python
def label(text_file):
    hot_array=[]
    file_name=text_file.split('.')[-3]
    if file_name=='vignere':
        hot_array=[0,1]
    if file_name=='simple_sub':
        hot_array=[1,0]
    return hot_array
```

```python
output=[]
hot_array=[]
for i in range(50):
    k=str(i)
    filepath=os.path.join("C:\\Users\\asd\\Desktop\\Dataset\\Vignere_dataset\\vignere. ("+k+").txt")
    text_file = open("C:\\Users\\asd\\Desktop\\Dataset\\Vignere_dataset\\vignere. ("+k+").txt", "r")
    lines = text_file.read()
    a=label(os.path.basename(filepath))
    hot_array.append(a)
    output.append(lines)
for l in range(50):
    k=str(l)
    filepath=os.path.join("C:\\Users\\asd\\Desktop\\Dataset\\Simple_sub_dataset\\simple_sub. ("+k+").txt")
    text_file = open("C:\\Users\\asd\\Desktop\\Dataset\\Simple_sub_dataset\\simple_sub. ("+k+").txt", "r")
    lines = text_file.read()
    a=label(os.path.basename(filepath))
    hot_array.append(a)
    output.append(lines)
b=shuffle(output,hot_array)
# print(hot_array)
```

```python
data_array=np.asarray(b[0])
label_array=np.asarray(b[1])
print(data_array)
```

```python
train_data, valid_data, train_label, valid_label = train_test_split(data_array_new, label_array, test_size=0.20)
print ("Training data",(np.shape(train_data)))
print ("Validiation data",(np.shape(valid_data)))
print(train_data[0])
```

```python
w1=np.random.normal(0, np.sqrt(0.05), (200,hidden))
#print w1
w2=np.random.normal(0, np.sqrt(0.1), (hidden,2))
#print w2
b1=np.zeros((hidden))
b2=np.zeros((2))
```

```python
def actFunc(haha,choose):
    if choose==1:#Sigmoid
        return (1/(1+np.exp(haha)))
    elif choose==2:#ReLU
        return np.max(0,haha)
    elif choose==3:#tanh
        return np.tanh(haha)
    elif choose==4:#softmax
        temp=np.exp(haha)
        return temp/np.sum(temp,axis=1,keepdims=True)
```

```python
def costFunc(y,t):
    return -(t*(np.log(y))+(1-t)*np.log(1-y))
```

```python
def forwardFunc(Data):
    a0=Data
    z1=np.dot(a0,w1)+b1
    a1=actFunc(z1,3)
    z2=np.dot(a1,w2)+b2
    a2=actFunc(z2,4)
    return a0,z1,a1,z2,a2
```

```python
def accuracyFunc(output,labelData,dataSize):
    maxInRow=np.argmax(output, axis=1)
    maxInLabels=np.argmax(labelData, axis=1)
    huehue=np.equal(maxInRow,maxInLabels)
    count=np.sum(huehue)
    return (count/dataSize)*100
```

```python
for j in range(10):
    #one epoch
    for i in range(80):#for each iteration, we use a for loop.
        a0,z1,a1,z2,a2=forwardFunc(train_data[i*1:(i+1)*1,:])
        y=a2
        labelBatch=train_label[i*1:(i+1)*1,:]
        #Backward pass
        del2=(y-labelBatch)
        del1=np.dot(del2,w2.T)*(1-pow(actFunc(z1,3),2))
        dcdw2=np.dot(a1.T,del2)
        dcdw1=np.dot(a0.T,del1)
        dcdb1=np.sum(del1,axis=0)
        dcdb2=np.sum(del2,axis=0)
        w1=w1-alpha*dcdw1
        w2=w2-alpha*dcdw2
        b2=b2-alpha*dcdb2
        b1=b1-alpha*dcdb1
    #Cost after one epoch
    a0,z1,a1,z2,a2=forwardFunc(train_data)
    ytrue=a2
    cost=np.sum(costFunc(ytrue,train_label))/80
    print("cost")
    print(cost)
    accuracy = accuracyFunc(ytrue,train_label,80)
    print("training accuracy")
    print(accuracy)
    ##validation
    Va0,Vz1,Va1,Vz2,Va2=forwardFunc(valid_data)
    Output=Va2
    acc = accuracyFunc(Output,valid_label,20)
    print("validation accuracy")
    print(acc)
```
